```python
import pandas as pd
import numpy as np
from IPython.display import display
def initializeDF(): #The file lmao2.csv is a downloaded excel sheet containing player information 
    df = pd.read_csv('lmao2.csv')
    df.drop(columns=['Timestamp', 'Discord tag', 'Preferred role:'], inplace=True)
    df.dropna(inplace=True)
    df['Battletag'].replace("[#].*", "", regex=True, inplace=True)
    df.columns = ["Name", "MMR", "Heros"]
    return df
    
#df = initializeDF() #Comment this line out to stop the 
df = df.drop(df[df["Name"]=="ESBN"].index)
display(df)
```
